```python
import pandas as pd
import matplotlib.pyplot as mat
import seaborn as sb
from sklearn import linear_model 
```

```python
trainDF = pd.read_csv("homeprices.csv")
trainDF
```

```python
sb.scatterplot(data=trainDF,x="area",y="price")
mat.show()
```

```python
reg_model = linear_model.LinearRegression()
reg_model.fit(trainDF[["area"]],trainDF.price)
```

```python
reg_model.predict([[3300]])
```

```python
myDF = pd.read_csv("areas.csv")
myDF
```

```python
predicted_price = reg_model.predict(myDF[["area"]])
myDF["predicted_price"]=predicted_price
myDF
```

```python
sb.lmplot(data=myDF,x="area",y="predicted_price")
mat.show()
```

```python
myDF.to_csv("final_result.csv",index=False)
```
