```python
import pandas as pd
import matplotlib.pyplot as mat
import seaborn as sb
from sklearn import linear_model
```

```python
myDF = pd.read_csv("insurance_data.csv")
myDF
```

```python
sb.regplot(data=myDF,x="age",y="bought_insurance",logistic=True)
mat.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(myDF[["age"]],myDF.bought_insurance,test_size=0.1)
```

```python
X_test
```

```python
reg_model = linear_model.LogisticRegression()
reg_model.fit(X_train,y_train)
```

```python
reg_model.predict(X_test)
```

```python
reg_model.predict([[25]])
```

```python
reg_model.predict([[40]])
```

```python
score = reg_model.score(X_test, y_test)
print("Model Accuracy:", score)
```
