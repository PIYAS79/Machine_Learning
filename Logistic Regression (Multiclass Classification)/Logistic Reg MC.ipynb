```python
import pandas as pd
import matplotlib.pyplot as mat
import seaborn as sb
from sklearn import linear_model
```

```python
trainDF = sb.load_dataset("iris")
trainDF
```

```python
trainDF["species"].value_counts()
```

```python
trainDF["species"].replace({'setosa':'1','versicolor':'2','virginica':'3'},inplace=True)
trainDF
```

```python
sb.countplot(data=trainDF,x="species",hue="species")
mat.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(trainDF[["sepal_length","sepal_width","petal_length","petal_width"]],trainDF.species,test_size=0.1)
```

```python
X_test
```

```python
reg_model = linear_model.LogisticRegression()
reg_model.fit(X_train,y_train)
```

```python
reg_model.predict(X_test)
```

```python
sb.pairplot(trainDF[["sepal_length","sepal_width","petal_length","petal_width","species"]],hue="species")
mat.show()
```

```python
reg_model.score(X_test,y_test)
```
