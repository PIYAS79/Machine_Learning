```python
from sklearn.datasets import make_regression
import numpy as np
import seaborn as sb
import matplotlib.pyplot as mat
```

```python
x,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
x
```

```python
y
```

```python
mat.scatter(x,y)
mat.show()
```

```python
# lets apply OLS (Ordinary List Square)
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y)
```

```python
# slop value (m)
reg.coef_
```

```python
# Intercept value (TARGET VALUE)
reg.intercept_
```

```python
# target value diye ekta line drow korlam, muloto amra ei line ta key pete chaitechi samne, 

mat.scatter(x,y)
mat.plot(x,reg.predict(x),color='red')
mat.show()
```

# Kaj ekhan theke suru :

#### m = 78.35
#### b = 0 [initial value hisebe dhore nilam]

```python
m = 78.35
b = 0
```

## ŷi = mXi+b

```python
ŷi = ((78.35*x)+0).reshape(4)
ŷi
```

```python
mat.scatter(x,y)
mat.plot(x,reg.predict(x),color='r',label="OLS")
mat.plot(x,ŷi,color='green',label='b = 0')
mat.legend()
mat.show()
```

## slope equ = -2 E(yi-mXi-b)

```python
loss_slope = -2 * np.sum(y - m * x.ravel() - b)
loss_slope
```

## η = 0.1 (ideal val 0.01 hoileo ami 0.1 dhorlam ar ki)

```python
η = 0.1
```

## step_size = η*slope

```python
step_size = η * loss_slope
step_size
```

```python
# calculate new "b"
b = b - step_size
b
```

```python
ŷi = ((m * x) + b ).reshape(4)
ŷi
```

```python
mat.scatter(x,y)
mat.plot(x,reg.predict(x),color="r",label="OLS")
mat.plot(x,ŷi,color="blue",label='b = 20.92')
mat.legend()
mat.title("kichu ta kacha kachi asche")
mat.show()
```

## iteration - 2 

```python
loss_slope = -2 * np.sum(y-m*x.ravel() -b)
loss_slope
```

```python
step_size = loss_slope * η
step_size
```

```python
b = b - step_size
b
```

```python
ŷi2 = ((m*x) + b).reshape(4)
ŷi2
```

```python
mat.scatter(x,y)
mat.plot(x,reg.predict(x),color="r",label="OLS")
mat.plot(x,ŷi2,color="blue",label='b = 20.92')
mat.legend()
mat.title("aro kichu ta kacha kachi asche")
mat.show()
```

## iteration - 3

```python
loss_slope = -2 * np.sum(y - m*x.ravel() -b)
loss_slope
```

```python
step_size = loss_slope * η
step_size
```

```python
b = b - step_size
b
```

```python
ŷi3 = ((m*x) + b).reshape(4)
ŷi3
```

```python
mat.scatter(x,y)
mat.plot(x,reg.predict(x),color="r",label="OLS")
mat.plot(x,ŷi3,color="blue",label='b = 20.92')
mat.legend()
mat.title("aro aro kichu ta kacha kachi asche")
mat.show()
```

## mane dekha gelo amra 3 ta iteration korei jaiga mton pouchaye gechi
## but ei kaj ta to amra raw vabe korlam, eitare jodi loop e kori , taile ekdom choto hoye jabe !!!

```python
b = 0
m = 78.35
η = 0.1
epochs = 10

for i in range(epochs) : 
    loss_slope = -2*np.sum(y-m*x.ravel()-b)
    b = b - (η*loss_slope)

    ŷi = m*x+b

    mat.plot(x,ŷi)
mat.scatter(x,y)
mat.show()
```

## η = learnig rate 
## learning rate muloto step gula re command kore, jodi aro choto nei taile amr step  gula aro choto hobe, graph er songkha bere jabe, iteration bere jabe 
## ar ete kore jodi epoch olpo hoi taile target porjonto na puchanor o ekta risk thake

## η = 0.01 dhore jodi kori taile jinish ta emn dekhabe

```python
b = 0
m = 78.35
η = 0.01
epochs = 10

for i in range(epochs) : 
    loss_slope = -2*np.sum(y-m*x.ravel()-b)
    b = b - (η*loss_slope)

    ŷi = m*x+b

    mat.plot(x,ŷi)
mat.scatter(x,y)
mat.show()
```
